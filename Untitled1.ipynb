```python
from transformers import pipeline
```

```python
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
```

```python
#EJEMPLO DE TEXTO
textos =[
    "Estoy muy feliz por cómo me salieorn las cosas.",
    "Me siento triste y demostivado.",
    "Tengo mucho miedo y no se que hacer.",
    "Estoy muy enojado  por lo que hiciste.",
    "Este dia es increible , todo me salio bien."]


 #CLASIFICACION DE LAS EMOCIONES POR CADA TEXTO
for texto in textos: # Removed extra spaces before this line
  resultado=classifier(texto)
  print(f"Texto:{texto}") # Changed 'Texto' to 'texto' to match the variable name
  for emocion in resultado[0]:
    print(f"{emocion['label']}:{emocion['score']:.4f}")
  print("\n") # Changed '/n' to '\n' for a newline chara
```
