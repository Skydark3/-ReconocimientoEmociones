```python
from transformers import pipeline
```

```python
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
```

```python
# Ejemplo de texto
textos = [
    "Estoy muy feliz por cómo salieron las cosas.",
    "Me siento triste y desmotivado.",
    "Tengo miedo de lo que pueda pasar.",
    "Estoy muy enojado por lo que hiciste.",
    "Este día es increíble, todo salió perfecto."
]

# Clasificación de emociones para cada texto
for texto in textos:
    resultado = classifier(texto)
    print(f"Texto: {texto}")
    for emocion in resultado[0]:
        print(f"{emocion['label']}: {emocion['score']:.2f}")
    print("\n")
```
